```python
# -*- coding: utf-8 -*-
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
%matplotlib inline

a = 7000000.
e = 0.2
theta = np.linspace(0, 2*np.pi, 500)
r = ((a*(1 - e**2)) / (1 + e * np.cos(theta)))/1000
R = 2*6521

U = u'\u2295'

linewidth=1.
interval=2000.

ax = plt.subplot(111, projection='polar')
ax.plot(theta, r, 'k', alpha=0.7, label='a = {:.3e} m; e = {:.2g}'.format(a,e))

ax.scatter(0., 0., s=(R), marker='.', facecolor='darkslategray', zorder=3, alpha=0.6)
ax.grid(True)
legend = ax.legend(loc='center left', shadow=True, bbox_to_anchor=(0.7, 1.15))
ax.set_title("ORBIT", va='bottom', y=1.08)
ax.set_rticks(np.delete(np.arange(0, ax.get_rmax()+1, interval), 0))
ax.set_rlabel_position(-45)
label_position=ax.get_rlabel_position()
ax.text(np.radians(label_position-5),ax.get_rmax()/2.,'Radius (km)',
        rotation=label_position,ha='center',va='center', size=9)
ax.text(0,0,U,ha='center',va='center', color='white',weight='bold', size=13)
ax.text(0.75*np.pi,np.max(r) * 1.5,'apoapsis {:.3e} km'.format(np.max(r)),
        rotation=0,ha='center',va='center', size=9)
plt.show()
```
